## Visualization 1


This interactive visualization shows an interactive scatter plot that represents humidity data over time. It shows the relationship between the columns "Min Humidity (%)" and "Max Humidity (%)" weekly for different years from 2018 to 2023. Users can select a specific year using the slider, and the scatter plot updates to show the data for the selected year. Additionally, there's a tooltip displays the currently selected year. When selecting a certain year, the space between the points might increase due to a higher dispersion or variability in the data with respect to the x & y axis.


**Horizontal Dispersion:**

If points horizontally are more dispersed, this suggests that the variables represented on the x-axis have a higher range ("Min Humidity").


**Vertical Dispersion:**

If points are vertically dispersed, this suggests that the variables represented on the y-axis have a higher range ("Max Humidity").

```python
import pandas as pd
import bqplot as bq
import ipywidgets as widgets
from IPython.display import display

# Calculate weekly averages
data['Year']=data['Date'].dt.year
data['Week'] = data['Date'].dt.isocalendar().week
weekly_averages = data.groupby(['Year', 'Week']).agg({
    'Min Humidity (%)': 'mean',
    'Max Humidity (%)': 'mean'
}).reset_index()

# Linear scales for x and y axes
x_scale = bq.LinearScale()
y_scale = bq.LinearScale()

# Scatter plot
scatter = bq.Scatter(
    x=weekly_averages['Min Humidity (%)'],
    y=weekly_averages['Max Humidity (%)'],
    scales={'x': x_scale, 'y': y_scale},
    default_size=100,
    interactions={'hover': 'tooltip'},
    display_legend=False,
)

# Tooltip
tooltip = bq.Tooltip(fields=['x', 'y'], formats=['.2f', '.2f'])
scatter.tooltip = tooltip

fig = bq.Figure(
    marks=[scatter],
    title='Relationship between Min and Max Humidity from 2018-2023',
    axes=[bq.Axis(scale=x_scale, label='Min Humidity (%)', orientation='horizontal'), 
          bq.Axis(scale=y_scale, label='Max Humidity (%)', orientation='vertical')],
)

# Slider for year selection
year_slider = widgets.IntSlider(
    value=data['Year'].min(),
    min=data['Year'].min(),
    max=data['Year'].max(),
    description='Year',
    continuous_update=False
)

# Function to update the scatter plot based on the selected year
def update_plot(change):
    selected_year = year_slider.value
    filtered_data = weekly_averages[weekly_averages['Year'] == selected_year]
    scatter.x = filtered_data['Min Humidity (%)']
    scatter.y = filtered_data['Max Humidity (%)']

# Attach the update_plot function to the slider's value change
year_slider.observe(update_plot, 'value')

# Display the components
display(widgets.VBox([year_slider, fig]))
```

```python
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import geopandas as gpd
from ipywidgets import interact  

# Assuming you have already defined telangana_gdf, district_mean_2022, and merged_df as mentioned in your code
def updatingplot(year):
    df_year = data[data['Year'] == year]#filtering by the selected year
    district_mean_year = df_year.groupby('District')['Min Wind Speed (Kmph)'].mean().reset_index()#grouping by the min wind speed by the mean
    merged_df_year = telangana_gdf.merge(district_mean_year, how='right', left_on='dtname', right_on='District')#right join
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))#creating the default plot
    merged_df_year.plot(column='Min Wind Speed (Kmph)', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)#adding the data to the plot
    ax.set_title(f'Choropleth Map of Min Wind Speed in Telangana ({year})')#setting title
    ax.set_axis_off()#turning off the axis labels
    plt.show()# show the map

# Create an interactive dropdown with years
interact(updatingplot, year=list(data['Year'].unique()));
```

## Visualization 3
### Weekly Minimum and Maximum temperatures from Jan 2018 to Jan 2023

This visualization aims to show the difference in mimimum and maximum temperatures from 2018-2023 on a weekly basis.

Interactivity shown: Slider for selecting the year range

Challenges: Intially i had wanted to plot this graph for the daily minimum and maximum temperatures but this resulted in crashing the dataset, I had then tried monthly as an alternative and found it to be sparse, and thus i had settled on creating a week-based visualization

```python
from ipywidgets import HTML

#find the min and max dates through all the years
min_date = data['Date'].min().year
max_date = data['Date'].max().year

#grouping the data weekly for minimum temp
weekly_min_temp= data.groupby(pd.Grouper(key='Date', freq='W'))['Min Temp (°C)'].mean()
weekly_min_temp= weekly_min_temp.reset_index()

#grouping the data weekly for maximum temp
weekly_max_temp=data.groupby(pd.Grouper(key='Date', freq='W'))['Max Temp (°C)'].mean()
weekly_max_temp=weekly_max_temp.reset_index()


def updateplot(year_range):
    start_year, end_year = year_range #defining the year range for the slider
    #filtering the min and max temps data based on the years selected
    selected_data_max = weekly_max_temp[(weekly_max_temp['Date'].dt.year >= start_year) & (weekly_max_temp['Date'].dt.year <= end_year)]
    selected_data_min = weekly_min_temp[(weekly_min_temp['Date'].dt.year >= start_year) & (weekly_min_temp['Date'].dt.year <= end_year)]
    #plotting the figure
    plt.figure(figsize=(14, 6))#defining the size
    plt.plot(selected_data_min['Date'], selected_data_min['Min Temp (°C)'], linestyle='-', label='Min Temp')
    plt.plot(selected_data_max['Date'], selected_data_max['Max Temp (°C)'], linestyle='-', label='Max Temp')
    #Setting the labels
    plt.xlabel('Year')
    plt.ylabel('Temperature (°C)')
    plt.title('Weekly Minimum and Maximum Temperatures from Jan 2018 - Oct 2023')
    plt.legend()
    plt.grid(True)
    plt.show()

#Using the intrangeslider for selecting the year
year_range_slider = widgets.IntRangeSlider(min=min_date, max=max_date, step=1,description='Select Year',value=[min_date, max_date])

#create the widget
widgets.interactive(updateplot, year_range=year_range_slider)
```

## Observation:

* Consistently, we can see that there is trend observed through the years for the min and max temperatures. Usually in the middle of the year, we can see the peaks in both min and max temperatatures.

* The observed lowest temperature is between 2022-2023 in the intial months of 2022

# Visualization 4

This visualization shows max temperature (in degrees celcius) over the given time frame for a combination of districts and mandals.  

This uses all data (2018-2023), but some combinations do not have updated data.

While making this visualization, I realized I needed to make it so not every mandal showed up for every district, only those that had data for that district.  

Aditionally, I wanted the user to be to quickly assess multiple different combinations so I added a "Random Combination" button. In making this button, I caused a problem that would make the mandal change twice before displayed, which I fixed.  

You can see in most combinations that there is a yearly pattern in max temperature; although, in 2018-2019, there seems to be some inconsistent data points.

```python
import pandas as pd
import bqplot.pyplot as plt
from bqplot import DateScale, LinearScale, Axis, Scatter, Figure
from ipywidgets import VBox, HBox, interactive_output, widgets
import random

# Create scales
x_sc = DateScale()
y_sc = LinearScale()

# Initial values for the interactive widgets
initial_district = "Adilabad"
initial_mandal = "Adilabad Rural"

# Filter data based on initial values
filtered_data = data[(data['District'] == initial_district) & (data['Mandal'] == initial_mandal)]

# Create scatter plot
scatter = Scatter(x=filtered_data['Date'], y=filtered_data['Max Temp (°C)'], scales={'x': x_sc, 'y': y_sc})

# Create axes
ax_x = Axis(label='Date', scale=x_sc, tick_format='%b %Y')
ax_y = Axis(label='Max Temp (°C)', scale=y_sc, orientation='vertical')

# Create figure
fig = Figure(marks=[scatter], axes=[ax_x, ax_y], title='Maximum Temperature by District and Mandal')

# Define update function
def update_district(change):
    # Update mandal options based on selected district
    mandal_dropdown.options = data[data['District'] == change.new]['Mandal'].unique()

def update(district, mandal):
    # Filter data based on selected district and mandal
    filtered_data = data[(data['District'] == district) & (data['Mandal'] == mandal)]

    # Update scatter plot data
    scatter.x = filtered_data['Date']
    scatter.y = filtered_data['Max Temp (°C)']

# Create interactive widgets
district_dropdown = widgets.Dropdown(options=data['District'].unique(), value=initial_district, description='District:')
mandal_dropdown = widgets.Dropdown(options=data['Mandal'].unique(), value=initial_mandal, description='Mandal:')
random_button = widgets.Button(description='Random Combination')

# Link the update_district function to the district dropdown's value
district_dropdown.observe(update_district, names='value')

# Define function for the random button click
def generate_random_combination(button):
    random_district = random.choice(data['District'].unique())
    random_mandal = random.choice(data[data['District'] == random_district]['Mandal'].unique())

    district_dropdown.value = random_district
    # mandal_dropdown.value = random_mandal
    # removed because multiple changes to mandal on click

# Link the button click event to the function
random_button.on_click(generate_random_combination)

# Create a spacer widget
spacer = widgets.HTML(value="<div style='width: 50px;'></div>")

# Create interactive output
out = interactive_output(update, {'district': district_dropdown, 'mandal': mandal_dropdown})

# Arrange widgets in a custom layout with spacer
controls = HBox([district_dropdown, mandal_dropdown, spacer, random_button])
display(VBox([controls, fig]))
```

## Visualization 5 
Total Rainfall in Telangana from 2018 - 2023. Takes the sum of the total rainfall per week to show the rainfall pattern.

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import widgets, interactive


# Get the min and max values from the 'Date' column
min_date = data['Date'].min().year
max_date = data['Date'].max().year

# Group by week and calculate the total rainfall like that
weekly_rainfall = data.groupby(pd.Grouper(key='Date', freq='W'))['Rain (mm)'].sum()
weekly_rainfall = weekly_rainfall.reset_index()

# Function to update the plot
def update_plot(year_range):
    start_year, end_year = year_range
    selected_data = weekly_rainfall[(weekly_rainfall['Date'].dt.year >= start_year) & (weekly_rainfall['Date'].dt.year <= end_year)]
    plt.figure(figsize=(14, 6))
    plt.plot(selected_data['Date'], selected_data['Rain (mm)'], marker='o', linestyle='-', color='blue', label='Weekly Rainfall')
    plt.xlabel('Date')
    plt.ylabel('Rainfall (mm)')
    plt.title(f'Total Weekly Rainfall from {start_year} to {end_year} in Telangana')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create a range slider widget for selecting the year range
year_range_slider = widgets.IntRangeSlider(
    min=min_date,
    max=max_date,
    step=1,
    description='Select Year Range',
    value=[min_date, max_date]
)

# Set up the interactive widget
interactive_plot = interactive(update_plot, year_range=year_range_slider)
interactive_plot
```
